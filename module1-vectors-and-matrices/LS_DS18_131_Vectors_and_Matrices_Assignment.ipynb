## 1.2 Create a three-dimensional vecor and plot it on a graph

```
vectors = np.array([[0,0,0,-1,-0,1]]) 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for vector in vectors: 
  v = np.array([vector[2],vector[3],vector[4]])
  vlength=np.linalg.norm(v)
  ax.quiver(vector[0], vector[1], vector[2], vector[3], vector[4], vector[5],        
              pivot='tail',length=vlength,arrow_length_ratio=0.3/vlength)
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.set_zlim([0,3]);
plt.title("Three-Dimensional Vector")
plt.show()â€©
```
